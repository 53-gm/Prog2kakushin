二変数関数のグラフを表示する用のプログラム

私は大学の解析学で二変数関数を勉強したが、グラフが想像もつかないので表示してみたいと思いこのプログラムを作成した。

RANGEの値を変えることで表示する範囲を変えることが可能
ROTATEの値を変えることでグラフを回転させることが可能

19行目のz = の部分を変えることで関数を表示する関数を変えることが可能

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

RANGE = (-3, 3) #関数を表示する範囲
ROTATE = (1, 30, 40) #グラフの回転 (0:回転させない 1:回転させる, 縦方向の角度(度数法), 横方向の角度(度数法)) 

#表示する領域の確保
fig = plt.figure(dpi=100)
ax = fig.add_subplot(projection='3d')

#ラベルの設定
ax.set_xlabel("x",size=10,color="black")
ax.set_ylabel("y",size=10,color="black")
ax.set_zlabel("z",size=10,color="black")

#x,yのデータ作成
data = np.linspace(RANGE[0],RANGE[1],100)
x, y = np.meshgrid(data,data)

# 表示したい関数を入力する↓
z = (x**2 + x**3 + y**2) * np.e ** (-x**2 -y **2)

#回転させる処理
if (ROTATE[0] == 1):
  ax.view_init(elev=ROTATE[1], azim=ROTATE[2])

#グラフ化
ax.plot_surface(x,y,z, cmap="jet")

plt.show()
```
