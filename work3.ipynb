コッホ曲線の描画

再帰的に関数kochを呼び出すことで歪な線が書けた。

参考:https://ja.wikipedia.org/wiki/%E3%82%B3%E3%83%83%E3%83%9B%E6%9B%B2%E7%B7%9A

```python
# turtle をColab環境で利用できるようにするため，Colabファイルを開いたときに一度だけ実行する
!pip3 install ColabTurtlePlus

# 各プログラムの先頭に以下のimport文を書いてください．
from ColabTurtlePlus.Turtle import *
```

```python
WINDOW_WIDTH = 300
WINDOW_HEIGHT = 900

def _setpos(t, pos): #線を描くやつの移動処理
  t.penup()
  t.setpos(pos)
  t.pendown()

def koch(t, len, n): #コッホ曲線を描く関数
  if n == 0:
    t.forward(len)
  else:
    koch(t, len / 3, n - 1)
    t.left(60) #左に60度回転
    koch(t, len / 3, n - 1)
    t.right(120) #右に120度回転
    koch(t, len / 3, n - 1)
    t.left(60) #左に60度回転
    koch(t, len / 3, n - 1)

clearscreen()   
setup(WINDOW_WIDTH,WINDOW_HEIGHT)

t = Turtle() #描画するものの作成
t.hideturtle() #線以外を非表示
t.speed(10) #描画スピード 0~10

for i in range(6):
  _setpos(t, (-WINDOW_WIDTH / 2, 400 - 150*i))
  koch(t, WINDOW_WIDTH, i)



```
